```python
def S(l,theta,phi):
    return ((np.sin(theta))**l)*(np.cos(l*phi))




THETA = np.linspace(-np.pi/2,np.pi/2,200)
PHI = np.linspace(0,2*np.pi,200)
theta,phi = np.meshgrid(THETA,PHI)

Z = S(0,theta,phi )
fig = plt.figure(figsize =(14, 9))

ax = plt.axes(projection='3d')

x = np.sin(phi) * np.cos(theta) * Z
y = np.sin(phi) * np.sin(theta) * Z
z = np.cos(phi) * Z


ax.plot_surface(x,y,z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title('s orbital')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.special as sp

# psi = radial(r,n,l)*angular(phi,theta,l,m)

def radial(r,n,l):
    a = np.sqrt((2/n)**3 * math.factorial(n-l-1) / (2*n*math.factorial(n+l)))
    b = np.exp(-r/n) * (2*r/n)**l
    c = sp.assoc_laguerre(2*r/n,n-l-1,2*l + 1)
    return a*b*c

def angular(phi,theta,l=0,m=0):
    a = sp.sph_harm(m,l,phi,theta)
    return a.real

r = np.linspace(0,100,100)
phi = np.linspace(0,np.pi,100)
theta = np.linspace(0,2*np.pi,100)
phi, theta = np.meshgrid(phi, theta)

n_max = 4

for n in range(1,n_max+1):
    for l in range(0,n):
        fig = plt.figure()
        plt.plot(r,radial(r,n,l)**2)
        plt.xlabel('r')
        plt.ylabel('Radial Function')
        plt.title('n='+str(n)+' l='+str(l))
        #plt.savefig('plots/R'+str(n)+str(l))
        for m in range(-l,l+1):
            angular_function = angular(theta,phi,l,m)
            x = np.sin(phi) * np.cos(theta) * abs(angular_function)
            y = np.sin(phi) * np.sin(theta) * abs(angular_function)
            z = np.cos(phi) * abs(angular_function)
            fig = plt.figure(figsize=(10,10))
            ax = fig.add_subplot(111, projection='3d')
            ax.plot_surface(x,y,z,alpha=0.5)
            ax.set_xlim(-0.75, 0.75)
            ax.set_ylim(-0.75, 0.75)
            ax.set_zlim(-0.75, 0.75)
            plt.title('n='+str(n)+' l='+str(l)+' m='+str(m))
            #plt.savefig('plots/Y'+str(n)+str(l)+str(m))
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.special as sp  


 


def wave_function(n, l, m, r, theta, phi):
    R = sp.genlaguerre(n - l - 1, 2 * l + 1)
    Y = sp.sph_harm(m, l, theta, phi)
    A = (((2 / n) ** 3) * (np.math.factorial(n - l - 1) / (2 * n * np.math.factorial(n + l))))
    B = np.exp(-r / n)
    C = (2 * r / n) ** l
    psi = (A * B * C) * R(dist) * Y

    return psi


for n in [1, 2, 3, 4] :
    for l in range(n):
        for m in range(-l, l + 1):
            
            X = np.linspace(-5 * n , 5 * n , 100) 
            Y = np.linspace(-5 * n , 5 * n, 100) 


            X, Y = np.meshgrid(X, Y)  
            Z = 1  

           
            dist = (X ** 2 + Y ** 2 + Z ** 2) ** 0.5  
            phi = np.arctan2(Y, X)  
            theta = np.arccos(Z / dist)  

            psi = np.abs(wave_function(n, l, m, dist, theta, phi))  

    
            fig = plt.figure(figsize=(10,10))
            ax = fig.add_subplot(111) 

            
            title = f'n={n}, l={l}, m={m}'  
            axes = ax.contourf(X, Y, psi, cmap='cool') 
            fig.colorbar(axes) 
            ax.set_title(title)
```
