# orbitals are plotted along z-axis in XY Plane

```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.special as sp  


 


def wave_function(n, l, m, r, theta, phi):
    R = sp.genlaguerre(n - l - 1, 2 * l + 1)
    Y = sp.sph_harm(m, l, theta, phi)
    A = (((2 / n) ** 3) * (np.math.factorial(n - l - 1) / (2 * n * np.math.factorial(n + l))))
    B = np.exp(-r / n)
    C = (2 * r / n) ** l
    psi = (A * B * C) * R(dist) * Y

    return psi


for n in [1, 2, 3, 4] :
    for l in range(n):
        for m in range(-l, l + 1):
            
            X = np.linspace(-5 * n , 5 * n , 100) 
            Y = np.linspace(-5 * n , 5 * n, 100) 


            X, Y = np.meshgrid(X, Y)  
            Z = 1  

           
            dist = (X ** 2 + Y ** 2 + Z ** 2) ** 0.5  
            phi = np.arctan2(Y, X)  
            theta = np.arccos(Z / dist)  

            psi = np.abs(wave_function(n, l, m, dist, theta, phi))  

    
            fig = plt.figure(figsize=(10,10))
            ax = fig.add_subplot(111) 

            
            title = f'n={n}, l={l}, m={m}'  
            axes = ax.contourf(X, Y, psi, cmap='cool') 
            fig.colorbar(axes) 
            ax.set_title(title) 
```
