```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
def S(r,theta):
    return ((np.sin(theta))/r)**2




r = np.linspace(1,5,50)
theta = np.linspace(-18,18,50)
r, theta = np.meshgrid(r, theta)

Z = S(r, theta )
fig = plt.figure(figsize =(14, 9))

ax = plt.axes(projection='3d')

ax.plot_surface(r, theta , Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_xlabel("r in meters")
ax.set_ylabel("theta in radian")
ax.set_zlabel("S")
ax.set_title('Q3: k*[sin(theta)/r]^2, where k=1, taken as constant')

                

```

```python
def S(r,theta):
    return ((np.sin(theta))/r)*np.cos(1-r)




r = np.linspace(-100,100,50)
theta = np.linspace(-18,18,50)
r, theta = np.meshgrid(r, theta)

Z = S(r, theta )
fig = plt.figure(figsize =(14, 9))

ax = plt.axes(projection='3d')

ax.plot_surface(r, theta , Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_xlabel("r in meters")
ax.set_ylabel("theta in radian")
ax.set_zlabel("E/B")
ax.set_title('Q5-b: k*sin(theta)*cos(1-r) /r, where k,w,c are taken as constant = 1')
```

```python
def f(t):
    return np.cos(t-1)
   

t = np.linspace(0,10,10)
plt.plot(t,f(t))
plt.xlabel('t in sec')
plt.ylabel("E/B")
plt.title('Q5(d): k*cos(t-1), where w,k,r,theta are taken as constant= 1')
```

```python
def f(w):
    return np.cos(w)
   

w = np.linspace(0,50,50)
plt.plot(w,f(w))
plt.xlabel("w is frequency")
plt.ylabel('E/B')
plt.title('Q5(d): k*cos(w), where t,k,r,theta are taken as constant= 1')
```
